Eliminaremos la columna "FAMI_TIENEINTERNET.1" ya que ya existe una y esta estaria duplicada

```python
data_clean = data.drop(columns=['FAMI_TIENEINTERNET.1'])
```

En la columna "ESTU_PRGM_ACADEMICO" hay algunos programas academicos que son tomados como diferentes categorias pero son lo mismo ya sea por una mayuscula, espacios o una tilde, lo que se hará es dejar todos en una categoria estandar para tener menos categorias diferentes.

```python
import unicodedata

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return ''.join([c for c in nfkd_form if not unicodedata.combining(c)])

data_clean['ESTU_PRGM_ACADEMICO'] = data_clean['ESTU_PRGM_ACADEMICO'].apply(lambda x: remove_accents(x.upper().strip()))

new_unique_programs = data_clean['ESTU_PRGM_ACADEMICO'].unique()
new_unique_programs_count = len(new_unique_programs)
new_unique_programs_example = new_unique_programs[:10]

new_unique_programs_count, new_unique_programs_example
```

Revisando la base de datos nos pudimos dar cuenta que habian varias carreras con errores de escritura que no sabiamos como arreglar por lo que toco hacer un trabajo "manual y a ojo" en donde se crearon dos funciones mientras revisavamos uno por uno, una para casos especificos (normalizar_carrera1) donde se queria cambiar nombres o errores de escritura de solo un programa academico y segundo esa ta funcion (normalizar_carrera) en donde ya mezclamos caracteristicas, por ejemplo si existia comunicacion y periodismo, comunicacion, periodismo; comunicacion-periodismo se ponen las dos caracteristicas en comun que son comunicacion y periodismo y se ejecuta la funcion para que todas queden con un mismo nombre

```python
def normalizar_carrera1(nombre):
    nombre = nombre.lower()
    if 'profesional en mercadeo'in nombre:
        return 'mercadeo'
    return nombre

data_clean['ESTU_PRGM_ACADEMICO'] = data_clean['ESTU_PRGM_ACADEMICO'].apply(normalizar_carrera1)
```

```python
def normalizar_carrera(nombre):
    nombre = nombre.lower()
    if 'ciencias' in nombre and 'deporte' and "educacion" in nombre:
        return 'ciencias del deporte y de la educacion fisica'
    return nombre

data_clean['ESTU_PRGM_ACADEMICO'] = data_clean['ESTU_PRGM_ACADEMICO'].apply(normalizar_carrera)
```

```python
new_unique_programs=data_clean['ESTU_PRGM_ACADEMICO'].unique()
new_unique_programs_count = len(new_unique_programs)
```

Importamos los nuevos datos limpios

```python
datos_nuevos= pd.read_csv('/content/datos1_train.csv')
datos_nuevos.info()
datos_nuevos.describe(include='all')
```

Queremos ver la correlacion que hay entre mis variables y la variable objetivo, como la mayoria son categoricas haremos prueba de chi cuadrado

```python
from scipy.stats import chi2_contingency

chi2_results = []

categorical_columns = datos_nuevos.columns[datos_nuevos.dtypes == 'object'].tolist()
categorical_columns.remove('RENDIMIENTO_GLOBAL')

for column in categorical_columns:
    contingency_table = pd.crosstab(datos_nuevos[column], datos_nuevos['RENDIMIENTO_GLOBAL'])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    chi2_results.append((column, chi2, p_value))

chi2_results_df = pd.DataFrame(chi2_results, columns=['Variable', 'Chi2', 'p-value'])
chi2_results_df.sort_values(by='p-value', inplace=True)

chi2_results_df
```

Como pudimos ver todas las variables son realmente importantes a la hora de querer predecir el rendimiento excepto ESTU_PRIVADO_LIBERTAD por lo que la eliminaremos de la base de datos

```python
datos_nuevos.drop(columns=['ESTU_PRIVADO_LIBERTAD'], inplace=True)
```

Como podemos ver hay varias columnas que solo tienes categorias como Si o No, por lo que para un modelo de machine learning es mas dificil entenderlo, lo que haremos es reemplazarlos por 0 y 1, sabiendo que 0 es No y 1 Si

```python
yes_no= {'Si': 1, 'No': 0}

columnas_a_convertir = ['FAMI_TIENECOMPUTADOR', 'ESTU_PAGOMATRICULAPROPIO', 'FAMI_TIENEAUTOMOVIL',
                      'FAMI_TIENELAVADORA', 'FAMI_TIENEINTERNET']

for column in columnas_a_convertir:
    datos_nuevos[column] = datos_nuevos[column].map(yes_no)

datos_nuevos.head()
```

```python
datos_nuevos['FAMI_ESTRATOVIVIENDA'].unique()
```

Llego la hora de revisar datos nulos o faltantes, revisamos la cantidad de datos faltantes y como todas las columnas que tienen datos faltantes son categoricas se decidio imputar con la moda.

```python
null_counts = datos_nuevos.isnull().sum()
null_counts = null_counts[null_counts > 0]

null_counts
```

```python
for column in null_counts.index:
    mode_value = datos_nuevos[column].mode()[0]
    datos_nuevos[column].fillna(mode_value, inplace=True)

remaining_nulls = datos_nuevos.isnull().sum().sum()
remaining_nulls
```

```python
datos_nuevos.info()
datos_nuevos['FAMI_ESTRATOVIVIENDA'].unique()
```

Ya vimos que no hay ningun datos faltante por lo que ya solo quedaria aplicarle one_hot_encoding a nuestra base de datos

Aca hacemos one hot encoding con la funcion get_dummies para poder tener los datos mejor y listos para meterlos al modelo.

```python
categorical_columns = datos_nuevos.select_dtypes(include=['object']).columns.drop('RENDIMIENTO_GLOBAL')

datos_nuevos = pd.get_dummies(datos_nuevos, columns=categorical_columns, drop_first=True)

datos_nuevos.info()
```

```python
datos_nuevos.head()
```

```python
for col in datos_nuevos.columns:
    if datos_nuevos[col].dtype == 'bool':
        datos_nuevos[col] = datos_nuevos[col].astype(int)
```
